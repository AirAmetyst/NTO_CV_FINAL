```python
import pandas as pd
import base64
from PIL import Image
import os
```

```python
df = pd.read_csv('data/EKB_images.csv', sep = ',')
df.head()
```

```python
TARGET_COLUMN = 'name'
```

```python
output_dir = "./data_by_name/"
os.makedirs(output_dir, exist_ok=True)
image_paths = []

for index, row in df.iterrows():
    image_data = base64.b64decode(row['img'])
    class_dir = os.path.join(output_dir, row[TARGET_COLUMN])
    os.makedirs(class_dir, exist_ok=True)
    image_path = os.path.join(class_dir, f"image_{index}.jpg")
    with open(image_path, "wb") as file:
        file.write(image_data)
    image_paths.append(image_path)

df['image_path'] = image_paths
```

```python
csv_file_path = 'processed_data_by_name.csv'
df.to_csv(csv_file_path, index=False)
```
